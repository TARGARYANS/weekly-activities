# NO:2 - HISTOGRAM

```python
plt.figure()
plt.hist(data1['PL'], rwidth = 0.5)
plt.title('DISTRIBUTION OF PL')
plt.xlabel('PL')
plt.ylabel('frequency')
plt.xticks
plt.yticks
```

```python
plt.figure()
plt.hist(data1['SL'], color = 'violet',rwidth = 0.5)
plt.title('DISTRIBUTION OF SL')
plt.xlabel('SL')
plt.ylabel('frequency')
plt.xticks
plt.yticks
```

# NO:3 - SCATTERPLOT

```python
plt.figure()
plt.scatter(data1['SW'], data1['PW'], s=5)
plt.xlabel('SW')
plt.ylabel('PW')
plt.title('PLOT OF SW & PW')
plt.xticks
plt.yticks
```
